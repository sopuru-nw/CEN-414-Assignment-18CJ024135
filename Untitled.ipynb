```python
data = pd.read_csv("UN_data.csv")
data2 = pd.read_csv("data_sorted.csv")
data2
```

```python
percentage = data[data["Country"].isin(countries)]

x = percentage['Country']
y = percentage['Value']

percentage.set_index('Country', inplace = True)
for year in range(1998,2007):
    try:
        percentage[percentage.Year == year].plot.barh(y='Value', title=year, figsize=(11,11))
        plt.title("Percentage of people working more than 40hrs per week in Africa")
        
    except:
        continue
print(percentage)

```

```python
col = ['1998','1999','2000','2001','2002','2003','2004','2005','2006']


plt.bar(col,inb[col].mean(), color='r')
plt.xlabel('years')
plt.ylabel('mean')
plt.title('Mean work hours')
```

```python
plt.bar(col,inb[col].std(), color='r')
plt.xlabel('years')
plt.ylabel('standard deviation')
plt.title('Standard deviation Working Hours')
```

```python
plt.bar(col,inb[col].median(), color='r')
plt.xlabel('years')
plt.ylabel('Median')
plt.title('Median Working Hours')
```

```python
_range = inb[col].max() - inb[col].min()
_range

iqrs =[]
for year in range(1998,2006):
    iqrs.append(inb[str(year)].quantile(0.75) - inb[str(year)].quantile(0.25))
    plt.ylabel("Interquatile range")
plt.xlabel("Years")
plt.title("INTERQUATILE RANGE INBOUND TOURISM")
plt.plot(range(2011,2019),iqrs, color='g')
```

```python
_range = data2[col].max() - data2[col].min()
_range.plot.line()
plt.ylabel("Range")
plt.xlabel("Years")
plt.title("RANGE OF AVERAGE WORKING HOURS")
```

```python

x = 1998
while x < 2007:   
    year = data[data.Year == x]
#     print(year)
    print('Year:', x)
    print('Mean:',year['Value'].mean())
    print('Standard Deviation:', year['Value'].std())
    print('Variance:',year['Value'].var())
    print('Median:',year['Value'].median())
    print('Mode:', year['Value'].mode())
    print('-----------------------------------')
    
    
    x += 1
```
